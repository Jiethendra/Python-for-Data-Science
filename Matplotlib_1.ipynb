```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
x=np.linspace(0,10,100)
fig=plt.figure
plt.plot(x,np.sin(x),"-")
plt.plot(x,np.cos(x),"--")
```

```python
y=np.linspace(0,10,100)
fig=plt.figure()
plt.plot(y,np.sin(x),"-")
plt.plot(y,np.cos(x),"--")
```

```python
plt.figure()
plt.subplot(2,1,1)
plt.plot(x,np.sin(x))

plt.subplot(2,1,2)
plt.plot(x,np.cos(x))
```

```python
fig, ax= plt.subplots(3)
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
ax[2].plot(x,np.sin(x))
```
